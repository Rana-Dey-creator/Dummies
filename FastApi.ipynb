```python
from fastapi import  FastAPI
from pydantic import BaseModel
import pickle
import json
import uvicorn
from pyngrok import ngrok
from fastapi.middleware.cors import CORSMiddleware
import nest_asyncio
```

```python
app = FastAPI()
```

```python
origins = ["*"]

app.add_middleware(
    CORSMiddleware,
    allow_origins=origins,
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)
```

```python
class model_input(BaseModel):

    pregnancies : int
    Glucose : int
    BloodPressure : int
    SkinThickness : int
    Insulin : int
    BMI : float
    DiabetesPedigreeFunction : float
    Age : int
```

```python
diabetes_model = pickle.load(open('/content/diabetes_model.sav', 'rb'))
```

```python
@app.post('/diabetes_prediction')
def diabetes_predd(input_parameters : model_input):

  input_data = input_parameters.json()
  input_dictionary = json.loads(input_data)

  preg = input_dictionary['pregnancies']
  glu = input_dictionary['Glucose']
  bp = input_dictionary['BloodPressure']
  skin = input_dictionary['SkinThickness']
  ins = input_dictionary['Insulin']
  bmi = input_dictionary['BMI']
  dpf = input_dictionary['DiabetesPedigreeFunction']
  age = input_dictionary['Age']

  input_list = [preg,glu,bp,skin,ins,bmi,dpf,age]

  prediction = diabetes_model.predict([input_list])

  if prediction[0] == 1:
    return 'The person is having diabetes'
  else:
    return 'The person is not having diabetes'
```

```python
ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)
nest_asyncio.apply()
uvicorn.run(app, port=8000)
```

```python
import pandas as pd

df = pd.read_csv('/content/diabetes.csv')
X = df.drop(columns=['Outcome']) # Features
y = df['Outcome']                # Target variable
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X, y) # This is where the "learning" happens
```

```python
import pickle

# Create the file and dump the model into it
with open('diabetes_model.sav', 'wb') as file:
    pickle.dump(model, file)
```

```python
# Load the model back into memory
loaded_model = pickle.load(open('diabetes_model.sav', 'rb'))

# Make a prediction with new data
result = loaded_model.predict([[5, 120, 70, 30, 0, 30.5, 0.5, 45]])
print(result)
```
